```python
#Bar Chart of GDP per Capita by Country
import matplotlib.pyplot as plt

country = ['UK', 'Pakistan', 'India', 'USA', 'UAE']
gdp_percapita = [46510, 1505, 2056, 83060, 50602]
plt.bar(country, gdp_percapita)
plt.title("Country vs GDP per Capita")
plt.xlabel("Country")
plt.ylabel("GDP per Capita")
plt.show()
```
